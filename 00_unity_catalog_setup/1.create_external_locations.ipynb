### Create external locations 
1. Bronze
2. Silver 
3. Gold

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS dbxexternal_imdbdev
 URL "abfss://imdbdev@dbximdbdata.dfs.core.windows.net/"
 WITH (STORAGE CREDENTIAL imdb_analytics);
```

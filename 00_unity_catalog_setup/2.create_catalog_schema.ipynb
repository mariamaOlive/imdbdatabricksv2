### Create Catalogs and Schemas 
1. Catalog - imdb_dev (Without managed location)
2. Schemas - bronze, silver, gold (With managed locations)

#### Creating Catalog

```sql
CREATE CATALOG IF NOT EXISTS imdb_dev
MANAGED LOCATION 'abfss://imdbdev@dbximdbdata.dfs.core.windows.net/';
```

```sql
USE CATALOG imdb_dev;
```

#### Creating Schemas

OBS: Creating a managed location for each layer just for learning purposed, otherwise inherit from the catalog.

```sql
CREATE SCHEMA IF NOT EXISTS bronze
MANAGED LOCATION 'abfss://bronze@dbximdbdata.dfs.core.windows.net/';
```

```sql
CREATE SCHEMA IF NOT EXISTS silver
MANAGED LOCATION 'abfss://silver@dbximdbdata.dfs.core.windows.net/';
```

### Create bronze tables (Extenal tables)
1. name.basics.tsv
2. title.basics.tsv
3. title.principals.tsv
4. title.ratings.tsv

##### 1. Load name.basics.tsv

```sql
CREATE TABLE IF NOT EXISTS imdb_dev.bronze.name_basics (
  nconst              STRING,
  primaryName         STRING,
  birthYear           INT,
  deathYear           INT,
  primaryProfession   STRING,
  knownForTitles      STRING
)
USING CSV
OPTIONS (
  path 'abfss://bronze@dbximdbdata.dfs.core.windows.net/name.basics.tsv',  
  header 'true',                               -- IMDb TSVs include a header row
  delimiter '\t',                              -- TSV
  nullValue '\N'                              -- IMDb uses \N for nulls
);
```

```sql
  SELECT * FROM imdb_dev.bronze.name_basics;
```

##### 2. Load title.basics.tsv

```sql
CREATE TABLE IF NOT EXISTS imdb_dev.bronze.title_basics (
  tconst          STRING,
  titleType       STRING,
  primaryTitle    STRING,
  originalTitle   STRING,
  isAdult         INT,        -- 0/1 in the IMDb TSV
  startYear       INT,
  endYear         INT,
  runtimeMinutes  INT,
  genres          STRING      -- comma-separated list
)
USING CSV
OPTIONS (
  path 'abfss://bronze@dbximdbdata.dfs.core.windows.net/title.basics.tsv',
  header 'true',
  delimiter '\t',
  nullValue '\N'
);
```

##### 3. Load title.principals.tsv

```sql
CREATE TABLE IF NOT EXISTS imdb_dev.bronze.title_principals (
  tconst      STRING,
  ordering    INT,
  nconst      STRING,
  category    STRING,   -- e.g., actor, director
  job         STRING,   -- sometimes \N
  characters  STRING    -- e.g., ["Character Name"]
)
USING CSV
OPTIONS (
  path 'abfss://bronze@dbximdbdata.dfs.core.windows.net/title.principals.tsv',
  header 'true',
  delimiter '\t',
  nullValue '\N'
);
```

##### 4. Load title.ratings.tsv

```sql
CREATE TABLE IF NOT EXISTS imdb_dev.bronze.title_ratings (
  tconst         STRING,
  averageRating  DOUBLE,
  numVotes       INT
)
USING CSV
OPTIONS (
  path 'abfss://bronze@dbximdbdata.dfs.core.windows.net/title.ratings.tsv',
  header 'true',
  delimiter '\t',
  nullValue '\N'
);
```

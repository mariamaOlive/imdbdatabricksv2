```python
from pyspark.sql.functions import current_timestamp
def add_ingestion_date(input_df):
  output_df = input_df.withColumn('ingestion_date', current_timestamp())
  return output_df
```

```python
from pyspark.sql.functions import col, when, lit, split
from pyspark.sql.types import ArrayType, StringType
from pyspark.sql.column import Column
from typing import Union
def split_to_array(col_or_name: Union[str, Column], sep: str = ",") -> Column:
    c = col(col_or_name) if isinstance(col_or_name, str) else col_or_name
    return when(c.isNull(), lit(None).cast(ArrayType(StringType()))) \
            .otherwise(split(c, sep))
```
